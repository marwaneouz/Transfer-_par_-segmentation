## Exercice 1 U-Net

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import ResNet50
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Configuration globale
IMG_HEIGHT = 256
IMG_WIDTH = 256
NUM_CLASSES = 91  # COCO dataset a 80 classes + background
BATCH_SIZE = 8
EPOCHS = 50

class AttentionBlock(layers.Layer):
    """Bloc d'attention pour améliorer U-Net"""
    def __init__(self, filters, **kwargs):
        super(AttentionBlock, self).__init__(**kwargs)
        self.filters = filters
        
        # Couches pour l'attention
        self.W_g = layers.Conv2D(filters, 1, padding='same')
        self.W_x = layers.Conv2D(filters, 1, padding='same')
        self.psi = layers.Conv2D(1, 1, padding='same', activation='sigmoid')
        self.relu = layers.ReLU()
        
    def call(self, g, x):
        """
        g: gating signal (du décodeur)
        x: feature map (de l'encodeur)
        """
        # Transformer les signaux
        g1 = self.W_g(g)
        x1 = self.W_x(x)
        
        # Addition et activation
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        
        # Appliquer l'attention
        return x * psi

def conv_block(inputs, filters, dropout_rate=0.1):
    """Bloc de convolution standard pour U-Net"""
    x = layers.Conv2D(filters, 3, activation='relu', padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(dropout_rate)(x)
    x = layers.Conv2D(filters, 3, activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    return x

def encoder_block(inputs, filters, pool=True, dropout_rate=0.1):
    """Bloc d'encodage avec pooling optionnel"""
    x = conv_block(inputs, filters, dropout_rate)
    if pool:
        p = layers.MaxPooling2D((2, 2))(x)
        return x, p
    return x

def decoder_block(inputs, skip_features, filters, use_attention=False, dropout_rate=0.1):
    """Bloc de décodage avec skip connections et attention optionnelle"""
    x = layers.Conv2DTranspose(filters, (2, 2), strides=2, padding='same')(inputs)
    
    if use_attention:
        # Appliquer l'attention aux skip connections
        attention = AttentionBlock(filters)
        skip_features = attention(x, skip_features)
    
    x = layers.concatenate([x, skip_features])
    x = conv_block(x, filters, dropout_rate)
    return x

def build_unet_basic(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)):
    """U-Net de base"""
    inputs = layers.Input(input_shape)
    
    # Encodeur
    s1, p1 = encoder_block(inputs, 64)
    s2, p2 = encoder_block(p1, 128)
    s3, p3 = encoder_block(p2, 256)
    s4, p4 = encoder_block(p3, 512)
    
    # Bridge
    b1 = conv_block(p4, 1024)
    
    # Décodeur
    d1 = decoder_block(b1, s4, 512)
    d2 = decoder_block(d1, s3, 256)
    d3 = decoder_block(d2, s2, 128)
    d4 = decoder_block(d3, s1, 64)
    
    # Sortie
    outputs = layers.Conv2D(NUM_CLASSES, 1, padding='same', activation='softmax')(d4)
    
    model = Model(inputs, outputs, name="U-Net_Basic")
    return model

def build_unet_with_attention(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)):
    """U-Net avec blocs d'attention"""
    inputs = layers.Input(input_shape)
    
    # Encodeur
    s1, p1 = encoder_block(inputs, 64)
    s2, p2 = encoder_block(p1, 128)
    s3, p3 = encoder_block(p2, 256)
    s4, p4 = encoder_block(p3, 512)
    
    # Bridge
    b1 = conv_block(p4, 1024)
    
    # Décodeur avec attention
    d1 = decoder_block(b1, s4, 512, use_attention=True)
    d2 = decoder_block(d1, s3, 256, use_attention=True)
    d3 = decoder_block(d2, s2, 128, use_attention=True)
    d4 = decoder_block(d3, s1, 64, use_attention=True)
    
    # Sortie
    outputs = layers.Conv2D(NUM_CLASSES, 1, padding='same', activation='softmax')(d4)
    
    model = Model(inputs, outputs, name="U-Net_Attention")
    return model

def build_unet_resnet_backbone(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)):
    """U-Net avec backbone ResNet50 pré-entraîné"""
    inputs = layers.Input(input_shape)
    
    # Backbone ResNet50 pré-entraîné (Transfer Learning)
    resnet = ResNet50(input_tensor=inputs, weights='imagenet', include_top=False)
    
    # Extraire les features maps à différentes résolutions
    # Noms des couches ResNet50 pour les skip connections
    skip_connection_names = [
        'conv1_relu',          # 128x128
        'conv2_block3_out',    # 64x64
        'conv3_block4_out',    # 32x32
        'conv4_block6_out'     # 16x16
    ]
    
    skip_connections = []
    for name in skip_connection_names:
        skip_connections.append(resnet.get_layer(name).output)
    
    # Bridge - sortie du ResNet
    bridge = resnet.output  # 8x8
    
    # Décodeur
    # Upsampling depuis 8x8 vers 16x16
    d1 = layers.Conv2DTranspose(512, (2, 2), strides=2, padding='same')(bridge)
    d1 = layers.concatenate([d1, skip_connections[3]])  # conv4_block6_out
    d1 = conv_block(d1, 512)
    
    # Upsampling depuis 16x16 vers 32x32
    d2 = layers.Conv2DTranspose(256, (2, 2), strides=2, padding='same')(d1)
    d2 = layers.concatenate([d2, skip_connections[2]])  # conv3_block4_out
    d2 = conv_block(d2, 256)
    
    # Upsampling depuis 32x32 vers 64x64
    d3 = layers.Conv2DTranspose(128, (2, 2), strides=2, padding='same')(d2)
    d3 = layers.concatenate([d3, skip_connections[1]])  # conv2_block3_out
    d3 = conv_block(d3, 128)
    
    # Upsampling depuis 64x64 vers 128x128
    d4 = layers.Conv2DTranspose(64, (2, 2), strides=2, padding='same')(d3)
    d4 = layers.concatenate([d4, skip_connections[0]])  # conv1_relu
    d4 = conv_block(d4, 64)
    
    # Upsampling final vers 256x256
    d5 = layers.Conv2DTranspose(32, (2, 2), strides=2, padding='same')(d4)
    d5 = conv_block(d5, 32)
    
    # Sortie
    outputs = layers.Conv2D(NUM_CLASSES, 1, padding='same', activation='softmax')(d5)
    
    model = Model(inputs, outputs, name="U-Net_ResNet50")
    return model

# Fonctions utilitaires pour l'entraînement

def dice_coefficient(y_true, y_pred, smooth=1e-6):
    """Coefficient de Dice pour évaluer la segmentation"""
    # Convertir les prédictions en classes prédites
    y_pred_classes = tf.argmax(y_pred, axis=-1)
    
    # S'assurer que les types sont corrects
    y_true = tf.cast(y_true, tf.int32)
    y_pred_classes = tf.cast(y_pred_classes, tf.int32)
    
    # Convertir en one-hot pour le calcul du Dice
    y_true_onehot = tf.one_hot(y_true, depth=NUM_CLASSES)
    y_pred_onehot = tf.one_hot(y_pred_classes, depth=NUM_CLASSES)
    
    # Aplatir les tenseurs
    y_true_f = tf.reshape(y_true_onehot, [-1, NUM_CLASSES])
    y_pred_f = tf.reshape(y_pred_onehot, [-1, NUM_CLASSES])
    
    intersection = tf.reduce_sum(y_true_f * y_pred_f, axis=0)
    union = tf.reduce_sum(y_true_f, axis=0) + tf.reduce_sum(y_pred_f, axis=0)
    dice = (2. * intersection + smooth) / (union + smooth)
    
    # Retourner la moyenne des classes (en excluant le background si nécessaire)
    return tf.reduce_mean(dice)

def dice_loss(y_true, y_pred):
    """Fonction de perte basée sur le coefficient de Dice"""
    return 1 - dice_coefficient(y_true, y_pred)

def iou_metric(y_true, y_pred):
    """Intersection over Union (IoU) metric"""
    # Convertir les prédictions en classes prédites
    y_pred_classes = tf.argmax(y_pred, axis=-1)
    
    # S'assurer que les types sont corrects
    y_true = tf.cast(y_true, tf.int32)
    y_pred_classes = tf.cast(y_pred_classes, tf.int32)
    
    # Convertir en one-hot
    y_true_onehot = tf.one_hot(y_true, depth=NUM_CLASSES)
    y_pred_onehot = tf.one_hot(y_pred_classes, depth=NUM_CLASSES)
    
    # Calculer l'IoU pour chaque classe
    intersection = tf.reduce_sum(y_true_onehot * y_pred_onehot, axis=[1, 2])
    union = tf.reduce_sum(y_true_onehot, axis=[1, 2]) + tf.reduce_sum(y_pred_onehot, axis=[1, 2]) - intersection
    
    iou = tf.reduce_mean((intersection + 1e-7) / (union + 1e-7))
    return iou

def create_data_generators(train_dir, val_dir):
    """Créer les générateurs de données pour l'entraînement"""
    # Générateur pour les images d'entraînement
    train_image_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        zoom_range=0.1
    )
    
    # Générateur pour les masques (pas d'augmentation pour préserver les labels)
    train_mask_datagen = ImageDataGenerator(rescale=1./255)
    
    # Générateurs de validation (pas d'augmentation)
    val_image_datagen = ImageDataGenerator(rescale=1./255)
    val_mask_datagen = ImageDataGenerator(rescale=1./255)
    
    # Note: Vous devrez adapter ces chemins selon votre structure de données COCO
    train_image_generator = train_image_datagen.flow_from_directory(
        train_dir +  r'C:\Users\user\Downloads\coco copy\val2017',
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode=None,
        seed=42
    )
    
    train_mask_generator = train_mask_datagen.flow_from_directory(
        train_dir + r'C:\Users\user\Downloads\coco copy\annotations_coco\instances_train2017.json',
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode=None,
        color_mode='grayscale',
        seed=42
    )
    
    return train_image_generator, train_mask_generator

def train_model(model, model_name, train_data=None, val_data=None):
    """Entraîner un modèle U-Net"""
    print(f"\n=== Entraînement de {model_name} ===")
    
    # Compiler le modèle
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-4),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy', dice_coefficient, iou_metric]
    )
    
    # Callbacks
    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7
        ),
        keras.callbacks.ModelCheckpoint(
            f'{model_name}_best.weights.h5',
            monitor='val_loss',
            save_best_only=True,
            save_weights_only=True
        )
    ]
    
    # Si vous avez des données réelles, utilisez-les ici
    if train_data is not None and val_data is not None:
        history = model.fit(
            train_data,
            validation_data=val_data,
            epochs=EPOCHS,
            callbacks=callbacks,
            verbose=1
        )
    else:
        # Données simulées pour démonstration
        print("Utilisation de données simulées pour la démonstration...")
        X_train = np.random.random((100, IMG_HEIGHT, IMG_WIDTH, 3))
        y_train = np.random.randint(0, NUM_CLASSES, (100, IMG_HEIGHT, IMG_WIDTH))
        X_val = np.random.random((20, IMG_HEIGHT, IMG_WIDTH, 3))
        y_val = np.random.randint(0, NUM_CLASSES, (20, IMG_HEIGHT, IMG_WIDTH))
        
        history = model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=5,  # Réduit pour la démonstration
            batch_size=BATCH_SIZE,
            callbacks=callbacks[:-1],  # Pas de sauvegarde pour données simulées
            verbose=1
        )
    
    return history

def plot_training_history(history, model_name):
    """Visualiser l'historique d'entraînement"""
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle(f'Historique d\'entraînement - {model_name}', fontsize=16)
    
    # Loss
    axes[0, 0].plot(history.history['loss'], label='Training Loss')
    axes[0, 0].plot(history.history['val_loss'], label='Validation Loss')
    axes[0, 0].set_title('Loss')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    
    # Accuracy
    axes[0, 1].plot(history.history['accuracy'], label='Training Accuracy')
    axes[0, 1].plot(history.history['val_accuracy'], label='Validation Accuracy')
    axes[0, 1].set_title('Accuracy')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Accuracy')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    
    # Dice Coefficient
    if 'dice_coefficient' in history.history:
        axes[1, 0].plot(history.history['dice_coefficient'], label='Training Dice')
        axes[1, 0].plot(history.history['val_dice_coefficient'], label='Validation Dice')
        axes[1, 0].set_title('Dice Coefficient')
        axes[1, 0].set_xlabel('Epoch')
        axes[1, 0].set_ylabel('Dice Score')
        axes[1, 0].legend()
        axes[1, 0].grid(True)
    
    # IoU
    if 'iou_metric' in history.history:
        axes[1, 1].plot(history.history['iou_metric'], label='Training IoU')
        axes[1, 1].plot(history.history['val_iou_metric'], label='Validation IoU')
        axes[1, 1].set_title('IoU Metric')
        axes[1, 1].set_xlabel('Epoch')
        axes[1, 1].set_ylabel('IoU Score')
        axes[1, 1].legend()
        axes[1, 1].grid(True)
    
    plt.tight_layout()
    plt.savefig(f'{model_name}_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

def visualize_predictions(model, test_images, test_masks, model_name, num_samples=3):
    """Visualiser les prédictions du modèle"""
    predictions = model.predict(test_images[:num_samples])
    predictions = np.argmax(predictions, axis=-1)
    
    fig, axes = plt.subplots(num_samples, 3, figsize=(12, 4*num_samples))
    fig.suptitle(f'Prédictions - {model_name}', fontsize=16)
    
    for i in range(num_samples):
        # Image originale
        axes[i, 0].imshow(test_images[i])
        axes[i, 0].set_title('Image originale')
        axes[i, 0].axis('off')
        
        # Masque de vérité terrain
        if test_masks is not None:
            axes[i, 1].imshow(test_masks[i].squeeze(), cmap='tab20')
            axes[i, 1].set_title('Vérité terrain')
            axes[i, 1].axis('off')
        
        # Prédiction
        axes[i, 2].imshow(predictions[i], cmap='tab20')
        axes[i, 2].set_title('Prédiction')
        axes[i, 2].axis('off')
    
    plt.tight_layout()
    plt.savefig(f'{model_name}_predictions.png', dpi=300, bbox_inches='tight')
    plt.show()

def main():
    """Fonction principale pour exécuter tous les modèles"""
    print("=== TP Transfer Learning - Segmentation d'images ===")
    print("Implémentation des architectures U-Net")
    
    # 1. U-Net de base
    print("\n1. Construction du U-Net de base...")
    unet_basic = build_unet_basic()
    unet_basic.summary()
    
    # Entraîner le modèle de base
    history_basic = train_model(unet_basic, "UNet_Basic")
    plot_training_history(history_basic, "UNet_Basic")
    
    # 2. U-Net avec attention
    print("\n2. Construction du U-Net avec attention...")
    unet_attention = build_unet_with_attention()
    unet_attention.summary()
    
    # Entraîner le modèle avec attention
    history_attention = train_model(unet_attention, "UNet_Attention")
    plot_training_history(history_attention, "UNet_Attention")
    
    # 3. U-Net avec backbone ResNet
    print("\n3. Construction du U-Net avec backbone ResNet50...")
    unet_resnet = build_unet_resnet_backbone()
    unet_resnet.summary()
    
    # Entraîner le modèle avec ResNet
    history_resnet = train_model(unet_resnet, "UNet_ResNet50")
    plot_training_history(history_resnet, "UNet_ResNet50")
    
    # Comparaison des performances
    print("\n=== Comparaison des performances ===")
    models_performance = {
        'UNet_Basic': history_basic.history,
        'UNet_Attention': history_attention.history,
        'UNet_ResNet50': history_resnet.history
    }
    
    # Graphique comparatif
    plt.figure(figsize=(15, 5))
    
    # Comparaison des loss de validation
    plt.subplot(1, 3, 1)
    for name, hist in models_performance.items():
        plt.plot(hist['val_loss'], label=name)
    plt.title('Validation Loss Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    # Comparaison des accuracy de validation
    plt.subplot(1, 3, 2)
    for name, hist in models_performance.items():
        plt.plot(hist['val_accuracy'], label=name)
    plt.title('Validation Accuracy Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    # Comparaison des Dice scores
    plt.subplot(1, 3, 3)
    for name, hist in models_performance.items():
        if 'val_dice_coefficient' in hist:
            plt.plot(hist['val_dice_coefficient'], label=name)
    plt.title('Validation Dice Score Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Dice Score')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('models_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Évaluation finale
    print("\n=== Résultats finaux ===")
    for name, hist in models_performance.items():
        final_val_loss = hist['val_loss'][-1]
        final_val_acc = hist['val_accuracy'][-1]
        print(f"{name}:")
        print(f"  - Validation Loss: {final_val_loss:.4f}")
        print(f"  - Validation Accuracy: {final_val_acc:.4f}")
        if 'val_dice_coefficient' in hist:
            final_dice = hist['val_dice_coefficient'][-1]
            print(f"  - Validation Dice: {final_dice:.4f}")
        print()

 
if __name__ == "__main__":
    # Vérifier la disponibilité du GPU
    print("GPUs disponibles:", tf.config.list_physical_devices('GPU'))
    
    # Exécuter le TP
    main()
     
```

```python
def aspp_block(inputs, filters=256, rate_scale=1):
    shape = inputs.shape
    b1 = layers.Conv2D(filters, 1, padding='same', use_bias=False)(inputs)
    b1 = layers.BatchNormalization()(b1)
    b1 = layers.Activation('relu')(b1)
    b2 = layers.Conv2D(filters, 3, padding='same', dilation_rate=6*rate_scale, use_bias=False)(inputs)
    b2 = layers.BatchNormalization()(b2)
    b2 = layers.Activation('relu')(b2)
    b3 = layers.Conv2D(filters, 3, padding='same', dilation_rate=12*rate_scale, use_bias=False)(inputs)
    b3 = layers.BatchNormalization()(b3)
    b3 = layers.Activation('relu')(b3)
    b4 = layers.Conv2D(filters, 3, padding='same', dilation_rate=18*rate_scale, use_bias=False)(inputs)
    b4 = layers.BatchNormalization()(b4)
    b4 = layers.Activation('relu')(b4)
    b5 = layers.GlobalAveragePooling2D()(inputs)
    b5 = layers.Reshape((1, 1, shape[-1]))(b5)
    b5 = layers.Conv2D(filters, 1, padding='same', use_bias=False)(b5)
    b5 = layers.BatchNormalization()(b5)
    b5 = layers.Activation('relu')(b5)
    b5 = layers.UpSampling2D((shape[1], shape[2]), interpolation='bilinear')(b5)
    x = layers.Concatenate()([b1, b2, b3, b4, b5])
    x = layers.Conv2D(filters, 1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    return x

def build_deeplabv3plus(input_shape, num_classes, backbone='xception', output_stride=16):
    inputs = layers.Input(shape=input_shape)
    if backbone == 'xception':
        base_model = tf.keras.applications.Xception(
            input_shape=input_shape,
            weights='imagenet',
            include_top=False
        )
        low_level_features = base_model.get_layer('block2_sepconv2_bn').output
        high_level_features = base_model.output
    x = aspp_block(high_level_features)
    x = layers.UpSampling2D((4, 4), interpolation='bilinear')(x)
    low_level = layers.Conv2D(48, 1, padding='same', use_bias=False)(low_level_features)
    low_level = layers.BatchNormalization()(low_level)
    low_level = layers.Activation('relu')(low_level)
    x = layers.Concatenate()([x, low_level])
    x = layers.Conv2D(256, 3, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(256, 3, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.UpSampling2D((4, 4), interpolation='bilinear')(x)
    outputs = layers.Conv2D(num_classes, 1, padding='same', activation='softmax')(x)
    model = Model(inputs, outputs, name='DeepLabV3+')
    return model




#   Entraînement :
 
model = build_deeplabv3plus(input_shape, num_classes, backbone='xception', output_stride=16)
model.compile(optimizer=tf.keras.optimizers.Adam(lr=1e-4), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(train_dataset, epochs=50, validation_data=val_dataset, callbacks=callbacks)



# Essayons ResNet101.

 
def build_deeplabv3plus_resnet(input_shape, num_classes):
    backbone = tf.keras.applications.ResNet101(
        input_shape=input_shape,
        weights='imagenet',
        include_top=False
    )
    inputs = backbone.input
    low_level_features = backbone.get_layer('conv2_block3_out').output
    high_level_features = backbone.output
    x = aspp_block(high_level_features)
    # Le reste est pareil
    return model


```

```python
# Code du GAN basique

def build_generator(latent_dim):
    noise = layers.Input(shape=(latent_dim,))
    x = layers.Dense(8*8*512, use_bias=False)(noise)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = layers.Reshape((8, 8, 512))(x)
    x = layers.Conv2DTranspose(256, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = layers.Conv2DTranspose(128, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    generated_image = layers.Conv2DTranspose(3, (5, 5), strides=(2, 2), padding='same', activation='tanh')(x)
    generator = Model(noise, generated_image, name='Generator')
    return generator

def build_discriminator(image_shape):
    image = layers.Input(shape=image_shape)
    x = layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same')(image)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same')(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Conv2D(256, (5, 5), strides=(2, 2), padding='same')(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Conv2D(512, (5, 5), strides=(2, 2), padding='same')(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Flatten()(x)
    validity = layers.Dense(1, activation='sigmoid')(x)
    discriminator = Model(image, validity, name='Discriminator')
    return discriminator

latent_dim = 100
image_shape = (128, 128, 3)
generator = build_generator(latent_dim)
discriminator = build_discriminator(image_shape)
discriminator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
noise = layers.Input(shape=(latent_dim,))
generated_image = generator(noise)
discriminator.trainable = False
validity = discriminator(generated_image)
gan = Model(noise, validity, name='GAN')
gan.compile(loss='binary_crossentropy', optimizer='adam')



#          GAN

def train_gan(generator, discriminator, gan, dataset, epochs, batch_size, latent_dim):
    for epoch in range(epochs):
        for batch in dataset:
            batch_size_actual = batch.shape[0]
            noise = np.random.normal(0, 1, (batch_size_actual, latent_dim))
            generated_images = generator.predict(noise)
            real_labels = np.ones((batch_size_actual, 1))
            fake_labels = np.zeros((batch_size_actual, 1))
            d_loss_real = discriminator.train_on_batch(batch, real_labels)
            d_loss_fake = discriminator.train_on_batch(generated_images, fake_labels)
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
            noise = np.random.normal(0, 1, (batch_size_actual, latent_dim))
            g_loss = gan.train_on_batch(noise, real_labels)
            if epoch % 100 == 0:
                print(f"Époque {epoch}: Perte D: {d_loss[0]:.4f}, Perte G: {g_loss:.4f}")
            if epoch % 500 == 0:
                save_generated_images(generator, epoch, latent_dim)

def save_generated_images(generator, epoch, latent_dim, examples=25):
    noise = np.random.normal(0, 1, (examples, latent_dim))
    generated_images = generator.predict(noise)
    generated_images = 0.5 * generated_images + 0.5
    fig, axs = plt.subplots(5, 5, figsize=(15, 15))
    cnt = 0
    for i in range(5):
        for j in range(5):
            axs[i,j].imshow(generated_images[cnt])
            axs[i,j].axis('off')
            cnt += 1
    fig.suptitle(f"Images créées - Époque {epoch}")
    plt.savefig(f"gen_epoch_{epoch}.png")
    plt.close()



#  Version avec attention
def self_attention(x):
    batch_size, height, width, channels = x.shape
    f = layers.Conv2D(channels // 8, 1)(x)
    g = layers.Conv2D(channels // 8, 1)(x)
    h = layers.Conv2D(channels, 1)(x)
    f_flat = layers.Reshape((-1, channels // 8))(f)
    g_flat = layers.Reshape((-1, channels // 8))(g)
    h_flat = layers.Reshape((-1, channels))(h)
    attention = tf.nn.softmax(tf.matmul(f_flat, g_flat, transpose_b=True))
    attended = tf.matmul(attention, h_flat)
    attended = layers.Reshape((height, width, channels))(attended)
    gamma = tf.Variable(0.0, trainable=True)
    output = gamma * attended + x
    return output

def build_sagan_generator(latent_dim):
    noise = layers.Input(shape=(latent_dim,))
    x = layers.Dense(4*4*1024, use_bias=False)(noise)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Reshape((4, 4, 1024))(x)
    x = layers.Conv2DTranspose(512, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Conv2DTranspose(256, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = self_attention(x)
    x = layers.Conv2DTranspose(128, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Conv2DTranspose(64, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = self_attention(x)
    x = layers.Conv2DTranspose(32, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    generated_image = layers.Conv2DTranspose(3, 3, strides=1, padding='same', activation='tanh')(x)
    generator = Model(noise, generated_image, name='SAGAN_Generator')
    return generator

def build_sagan_discriminator(image_shape):
    image = layers.Input(shape=image_shape)
    x = layers.Conv2D(64, 4, strides=2, padding='same')(image)
    x = layers.LeakyReLU(0.1)(x)
    x = layers.Conv2D(128, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.1)(x)
    x = self_attention(x)
    x = layers.Conv2D(256, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.1)(x)
    x = layers.Conv2D(512, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.1)(x)
    x = self_attention(x)
    x = layers.Conv2D(1024, 4, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.1)(x)
    x = layers.GlobalAveragePooling2D()(x)
    validity = layers.Dense(1, activation='sigmoid')(x)
    discriminator = Model(image, validity, name='SAGAN_Discriminator')
    return discriminator
```
